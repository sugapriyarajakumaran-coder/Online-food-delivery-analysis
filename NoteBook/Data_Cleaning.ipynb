```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"], errors='coerce')
```

```python
df["City"].unique()
```

```python
cat_columns = ["Customer_Gender",
    "City",
    "Area",
    "Cuisine_Type",
    "Payment_Mode",
    "Order_Status",
    "Cancellation_Reason",
    "Peak_Hour",
    "Order_Day"]

for col in cat_columns:
    df[col] = df[col].astype('category')

print(df.dtypes)    
```

```python
df.isnull().sum()
```

```python
null_percent = df.isnull().mean() * 100
print(null_percent.sort_values(ascending=False))
```

```python
df["Cancellation_Reason"] = df["Cancellation_Reason"].astype("object")
df["Cancellation_Reason"] = df["Cancellation_Reason"].fillna("Not Cancelled") 
df["Cancellation_Reason"] = df["Cancellation_Reason"].astype("category") 
df["Cancellation_Reason"].value_counts()
```

```python
df["Customer_Gender"] = df["Customer_Gender"].astype("object")
df["Customer_Gender"] = df["Customer_Gender"].fillna("Unknown") 
df["Customer_Gender"] = df["Customer_Gender"].astype("category") 
df["Customer_Gender"].value_counts()
```

```python
df["Customer_Age"] = df["Customer_Age"].fillna(df["Customer_Age"].mean())
```

```python
df.drop("Order_Time", axis=1, inplace=True)
```

```python
df["Order_Date"] = df["Order_Date"].fillna(df["Order_Date"].mode()[0])
df["Discount_Applied"] = df["Discount_Applied"].fillna(0)
df["Order_Value"] = df["Order_Value"].fillna(df["Order_Value"].mean())
```

```python
df["Delivery_Time_Min"] = df["Delivery_Time_Min"].fillna(df["Delivery_Time_Min"].mean())
df["Delivery_Time_Min"].skew()
```

```python
df["Distance_km"] = df["Distance_km"].fillna(df["Distance_km"].median())
```

```python
df['Delivery_Rating'] = df['Delivery_Rating'].fillna(df['Delivery_Rating'].median())
df['Final_Amount'] = df['Final_Amount'].fillna(df['Order_Value'] - df['Discount_Applied'])

```

```python
cat_cols = ['Cuisine_Type', 'City', 'Area','Peak_Hour','Payment_Mode']

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
import pandas as pd

numeric_cols = ['Delivery_Time_Min', 'Order_Value']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    

    df[col] = df[col].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

    print(f"{col} capped: lower={lower_bound:.2f}, upper={upper_bound:.2f}")
```

```python
df['Delivery_Rating'] = df['Delivery_Rating'].clip(upper=5)
df['Profit_Margin'] = df['Profit_Margin'].clip(lower=0)
```

```python
cat_cols = df.select_dtypes(include='category').columns
print(cat_cols)

for col in cat_cols:
    df[col] = (df[col].astype(str).str.strip().str.lower().astype('category'))

print("Categorical columns standardized successfully!")
```
